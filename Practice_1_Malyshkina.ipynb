```python
class Chocolate:
  def __init__(self, price: float, filling: str = ''):
    self.filling = filling
    if price > 0:
      self.price = price
    else:
      raise ValueError('Unacceptable price given')

  def __str__(self):
    return f'Шоколадный батончик, {self.price} рублей'

  def show_my_chocolate(self):
    if self.filling == '':
      return f'Обычный шоколадный батончик'
    else:
      return f'Шоколадный батончик, добавка: {self.filling}'
```

```python
MomsFavourite = Chocolate(100.5, 'Salted caramel')
DadsFavourite = Chocolate(74.03, 'Hazelnut')
SomeonesFavourite = Chocolate(9, 'Mystery filling')
GransFavourite = Chocolate(35)
MyFavourite = Chocolate(16, 'Pistachio')
```

```python
print(GransFavourite.show_my_chocolate())
```
